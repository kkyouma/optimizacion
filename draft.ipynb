## Ejemplo 2: Fuerza bruta

La empresa minera American Minerals Ltda esta explorando un nuevo yacimiento en una region rica en recursos minerales, tales como cobre, oro y plata. La empresa debe determinar la canitidad optima por medio de estudios descubrio que los ingresos diaros (en miles de dolares) obtenidos por la venta de mineral en funcion de la cantidad diaria extraida.

$$f(x) = -12(x-20)^3$$

Donde, x representa la cantidad de mineral extraido por dia, medida en toneladas.

*Determinar el valor optimo por medio de Fuerza bruta en Excel y Python*

```python
def obj_function(x):
    return -12*x*(x-20)**3
```

```python
start = 0
stop = 20
num = 100
x = np.linspace(start, stop, num)
y = obj_function(x)

value = -np.inf
for e in x:
    val_fun = obj_function(e)
    if val_fun > value:
        value = val_fun
        parameters = [round(e), round(val_fun)]
print(f"Solution: x = {parameters[0]}, y = {parameters[1]}")
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8))
axs[0].plot(x, y, color="black", label="$f(x) = -12(x-20)^3$")
axs[0].set_xlabel("Cantidad de mineral en toneladas extraido por dia (ton)")
axs[0].set_ylabel("Ingresos diaros (miles de dolares)")
axs[0].set_title("Funcion de ingresos diarios y punto optimo de maximizacion")

axs[1].plot(x, y, color="black")
axs[1].scatter(parameters[0], parameters[1], s=300, color="red")
axs[1].grid(True)
```
