```python
def obj_function(x):
    return 3 * x * (x - 20)**3


start = -5
stop = 30
num = 100

x = np.linspace(start, stop, num)
y = obj_function(x)
zero_line = np.zeros_like(x)

plt.plot(x, y)
plt.plot(x, zero_line, color="black", linestyle="--", alpha=0.4)
plt.show()
```

```python
import sympy as sp

x = sp.Symbol('x')

x0 = 40
TOL = 1e-6
MAX_ITER = 10000

def obj_function(x):
    return 3 * x * (x - 20)**3

def get_derivative():
    return sp.diff(obj_function(x), x)

def get_newton_optimization(x0, TOL, MAX_ITER):
    dfx0 = get_derivative()  
    fx0 = obj_function()
    count = 0
    for _ in range(MAX_ITER):
        count += 1  
        x1 = x0 - float(fx0(x0)/dfx0.subs(x, x0))
        
        if abs(x1 - x0) < TOL:
            break
        
        x0 = x1 
        
    return (x1, count)

root, iterations = get_newton_optimization(x0, TOL, MAX_ITER)
print(f"La raíz es: {root} en {iterations} iteraciones")
```
