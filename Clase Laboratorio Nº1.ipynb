    <br>

```python
# .append()--> se usa para agregar un elemento al final de una lista.

a.append(12) 
print("a: ", a)
```

```python
#len() --> entrega la cantidad de elementos que contiene una lista

print("Largo de b: ", len(b)) 
```

```python
# sorted() --> ordena la listas

print("Lista ordenada: ", sorted(lista_desordenada)) #ordena la lista de menor a mayor
print("Lista ordenada al reves: ", sorted(lista_desordenada, reverse=True)) # ordena la lista de mayor a menor.
```

```python
# .clear() --> elimina todos los elementos de la lista.

lista_desordenada.clear()
print(lista_desordenada, end ="\n" )
```

```python
# .count() --> arroja el número de elementos con el valor indicado.
print(b.count(1.6))   
```

```python
#.extend() --> añade los elementos de una lista (o cualquier iterador) al final de la lista actual.

lista1 = [1, 2, 3]
lista2 = [4, 5, 6]
lista1.extend(lista2)
print(lista1)
```

```python
#.insert() --> añade un elemento en la posición que se indica.


numeros = [1, 2, 3, 4, 5]
numeros.insert(2, 10) 
print(numeros)
```

```python
# del  --> eliminar elementos por índice.
lista = [1, 2, 3]

del lista[2]
print(lista)
```

```python
#.pop() --> elimina el elemento de la posición que se indica.

letras = ['a', 'b', 'c', 'd']
elemento_eliminado = letras.pop(2) 
print(letras)
print("Elemento eliminado:", elemento_eliminado)
```

```python
#.reverse() --> invierte el orden de la lista.
lista = [1, 2, 3, 4, 5]
lista.reverse()
print(lista)
```

```python
#.set() --> elimina los duplicados de la lista.

duplicados = [3, 1, 4, 1]
sin_duplicados = list(set(duplicados))
print("Lista sin duplicados:", sin_duplicados)
```

```python
# Operaciones 
a =[2,5,6,8]
b =[1, 4,5.5,7]

print("Suma de listas: ", a+b, end= "\n\n")
# Nota: La resta, multiplicacion y la division de listas no esta permitida -> a-b, a*b, a/b.

#print("Lista:", b, end="\n\n")
print(b[:])  #devuelve una copia toda la lista.
print(b[1:]) #devuelve la lista desde el segundo elemento hasta el final
print(b[:2],end="\n\n\n") #devuelve la lista desde el principio hasta el tercer elemento (sin incluir el índice 2)


# Recorrer una lista y agregar un valor al final
cont = 100
for i in range(len(b)):
    a.append(cont)
    cont+=1
    print(a)

    
```

```python
# Ejemplo: Eliminar los elementos negativos de una lista, sustituyéndolos por 0 y los 0 cambiarlos 8. 
b = [1, 3, -6, 5, -1, 3, -15, -100, 77,0,0,-2]

for i in range(len(b)):
    if b[i] == 0:
        b[i] =8
    if b[i] < 0: 
        b[i] = 0
    

print(b)
```

```python
#zip() --> Crea una matriz a partir de dos lista de elementos
nombres = ["Evelyn", "Kara", "Anna",
           "Camila", "Daniel", "James",
           "Jaime", "Clair", "Marta", 
           "Mike", "Katherine", "Fernanda"]


telefonos = []
for i in range(len(nombres)):
    telefonos.append(random.randint(9, 999999999)) # para generar un número aleatorio entero en el rango desde 9 hasta 999999999 (inclusive)

dic = dict(zip(nombres,telefonos))
print(dic)
```

```python
# Añadir elementos
# Nota: los diccionarios solo pueden añadir elementos indicando el nombre del diccionario,
#       la nueva clave y su valor.

dic["Lauren"] = 951384738
print(dic)
```

```python
# Eliminar valores de un diccionario

del dic["Marta"]
print(dic)
```

```python
# Recorrer un diccionario
# Formas de obtener las claves
for keys in dic:
    print(dic[keys]) 

print(" ")

for valores in dic.values():
    print(valores)
```

```python
# Forma de obtener los valores
for valor in dic.keys():
    print(valor)
```

```python
#items() --> Devuelve una lista de tuplas, cada tupla se compone de dos elementos: el primero será la clave y el segundo, su valor.
print(dic.items())
```

```python
print(dic.pop("Daniel"), end="\n\n")
print(dic)
```

```python
x = [1,3,5,7,19,20,40,41,50]
y = [2,4,6,8,20,25,36,49,50]

plt.figure(figsize=(8, 6)) 
plt.rcParams['font.family'] = 'Times New Roman'

plt.plot(x, y, marker='o', color='blue', linestyle='-', label='Datos Listas', linewidth=3) 
plt.xlabel("Eje x", fontsize=25, fontweight="bold") 
plt.ylabel("Eje y", fontsize=25, fontweight="bold")
plt.title('Gráfico a partir de listas', fontsize=25, fontweight="bold")
plt.legend(fontsize=25)  # Mostramos la leyenda
plt.grid(True)  # Generamos cuadrícula
plt.tick_params(axis='both', which='major', labelsize=25, length = 18, color = "black")
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)  # Utilizamos numpy para generar valores de -10 a 10 
                               # con 100 puntos intermedios para el eje x.

# Calcular los valores de y para crear una parábola y = x**2
y = x ** 2

plt.figure(figsize=(8, 6))  # Definimos el tamaño del gráfico
plt.rcParams['font.family'] = 'Times New Roman' # Cambiamos el tipo de letra

plt.plot(x, y, label = "y = $x^2$", color = "red",linewidth=3)  # Dibujar la parábola
plt.plot(x+3, y+3,"-.", label = "y = $x^2+3$", color = "green", linewidth=3)

plt.xlabel('Eje x', fontsize=25, fontweight="bold")  # Generamos la etiqueta para el eje x
plt.ylabel('Eje y', fontsize=25, fontweight="bold")  # Generamos la etiqueta para el eje y

plt.title('Gráfico a partir de una función Numpy', fontsize=25, fontweight="bold")  # Generamos la etiqueta del título del gráfico
plt.legend(fontsize=25)  # Mostramos la leyenda
plt.grid(True)  # Generamos cuadrícula
plt.tick_params(axis='both', which='major', labelsize=25, length = 18, color = "black") # Modificamos el tamaño de los ejes
plt.tight_layout()  # ajusta el tamaño del gráfico
plt.show()  # Mostramos el gráfico
```
